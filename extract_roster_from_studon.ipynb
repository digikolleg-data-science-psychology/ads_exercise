```python
import pandas as pd
from pathlib import Path
import glob
import os
import time
```

### Extract student identifiers from StudOn export for import in GitHub Classroom

```python
# change this to the current semester
semester = "SoSe_24"
```

```python
studon_path = Path("studon_export")
roster_path = Path("student_roster") 
```

```python
list_of_files = glob.glob("studon_export/*.csv")
latest_export = max(list_of_files, key=os.path.getctime) # get the latest export
latest_export
```

```python
roster_df = pd.read_csv(latest_export)
roster_df.head()
```

```python
idm_identifiers = roster_df["Login"]
idm_identifiers
```

```python
# export csv
timestr = time.strftime("%Y%m%d-%H%M%S")
idm_identifiers.to_csv(roster_path.joinpath(semester+"-"+timestr), index=False, header=False)
```
