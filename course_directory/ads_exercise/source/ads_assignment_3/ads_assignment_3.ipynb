#### 3. Datatypes - List
A Python list is a structure that allows you to store a collection of elements of any data type in a single variable, like

| functionality | key word|explanation | example | return |

##### 4.2 Implement a function that checks the potassium level of a patient. The potassium level is usually between 3.7 and 5.3 mEq/l (milliequivalents per liter). To do so:
* create a function called ``check_potassium`` that takes ``value`` as an argument
* check if ``value`` is in the given normal potassium range
    * if yes: print an informative message (e.g. "the potassium level is in the correct range") and return False
    * if no: print an informative message and return True
* Play around with different value to see if your implementation works correctly

**Hint:** You can combine two conditions using logical operators.

````python
### BEGIN SOLUTION
def check_potassium(value):
    further_actions = False
    if value >= 3.7 and value <= 5.3:
        print("the potassium level is in the correct range")
    else:
        further_actions = True
        print("the potassium level is out of range")
    return further_actions
### END SOLUTION
````

##### 4.3 Implement a function that checks the hemoglobin level based on the sex of the patient. The hemoglobin level of females is usually between 12.1 and 15.1 g/dl (grams per deciliter), for males it lays between 13.8 and 17.2 g/dl. To do so:
* create a function called ``check_hemoglobin`` that takes ``sex`` and ``value`` as arguments
* create a nested if-construct
    * outer if-statement checks if ``sex`` is female or male
    * inner if-statements checks based on the sex the correct range of hemoglobin level
        * if it is in range print an informative message and return False
        * if it is **not** in the correct range print an informative message and return True



````python
### BEGIN SOLUTION
def check_hemoglobin(sex, value):
    further_actions = False
    if sex == "female":
        if value >= 12.1 and value <= 15.1:
            print("the hemoglobin level is in the correct range")
        else:
            further_actions = True
            print("the hemoglobin level is out of range")
    elif sex == "male":
        if value >= 13.8 and value <= 17.2:
            print("the hemoglobin level is in the correct range")
        else:
            further_actions = True
            print("the hemoglobin level is out of range")
    return further_actions
### END SOLUTION
````

````python
assert check_hemoglobin("female", 16.0) == True
assert check_hemoglobin("male", 16.0) == False
### BEGIN HIDDEN TESTS
assert check_hemoglobin("female", 13.0) == False
assert check_hemoglobin("female", 12.0) == True
assert check_hemoglobin("male", 13.0) == True
assert check_hemoglobin("male", 18.0) == True
### END HIDDEN TESTS
````
