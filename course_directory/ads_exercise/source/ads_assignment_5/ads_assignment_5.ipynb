#### 2. Datatypes - Dictionary
A dictionary is a type of data structure that stores key-value pairs. Each key in the dictionary is associated with a value:
```
dict_1 = {
    key_1: value_1,
    key_2: value_2
}
```
While dictionary keys must be unique, dictionary values can be duplicated. To access the value associated with a particular key, you can use square brackets:
```
dict_1[key_1]
```
##### 2.1 Create a dictionary called ``patient`` with the following key-value pairs:

|name|Jane|string|

|allergies|pollen, peanuts|list|
|blood_test_results|None|Null|

````python
### BEGIN SOLUTION
patient = {
    "name": "Jane",
    "height": 167.3,
    "vegetarian": True,
    "allergies": ["pollen", "peanuts"],
    "blood_test_results": None
}
### END SOLUTION
````
