### **Jupyter (Colaboratory) Notebooks**
A Jupyter/Colaboratory Notebook is an interactive tool that allows you to take notes, execute code, create figures, and easily share your work with other people. It always has the file extension .ipynb which stands for **i**nteractive **Py**thon **n**ote**b**ook. Notebooks comprise a series of cells. Each cell (looks like a rectangular block of text) in a notebook contains one of two basic categories of content:
1. **Markdown cells**: contains formatted text. We use this type of cell to give explanations and instructions about the assignment tasks.
2. **Code cells**: contains computer code (in our case Python code) which is a set of instructions that tell the notebook what to do when the cell is *run*. To execute/run the code in a cell hold the ``shift`` key and press ``enter/return`` or press the start button in the upper left corner of the cell.
