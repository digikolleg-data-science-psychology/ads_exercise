```python
assignment_list = ["Assignment_1", "Assignment_2", "Assignment_3", "Assignment_4", "Assignment_5"]
```

```python
total = {}
```

```python
for assignment in assignment_list:
    df = pd.read_excel(exercise_path / f"{assignment}.xlsx")

    print(assignment)

    try:
        assignment_df = df[["Login", 1]]
    except KeyError:
        assignment_df = df[["Login", "Test Results in Marks"]]
    assignment_df.columns = ["idm", "passed"]
    assignment_df.set_index("idm", inplace=True)
    assignment_df.replace("passed", "Passed", inplace=True)

    assignment_df["passed"] = assignment_df["passed"] == "Passed"

    total[assignment] = assignment_df
```
