```python
import pandas as pd
from pathlib import Path
```

```python
exercise_path = Path("/Users/abelluc/FAUbox/DigiKolleg/BPOS/Exercise/WS24_25_Results")
```

```python
# Read in the data
```

```python
assignment_list = ["Assignment_1", "Assignment_2", "Assignment_3", "Assignment_4", "Assignment_5", "Assignment_6"]
```

```python
total = {}
```

```python
for assignment in assignment_list:
    df = pd.read_excel(exercise_path / f"{assignment}.xlsx")

    print(assignment)

    try:
        assignment_df = df[["Login", 1]]
    except KeyError:
        assignment_df = df[["Login", "Test Results in Marks"]]
    assignment_df.columns = ["idm", "passed"]
    assignment_df.set_index("idm", inplace=True)
    assignment_df.replace("passed", "Passed", inplace=True)

    assignment_df["passed"] = assignment_df["passed"] == "Passed"

    total[assignment] = assignment_df
```

```python
total_df = pd.concat(total)
```

```python
# convert dict to dataframe
total_df.index.names = ["assignment", "idm"]
total_df.sort_index(inplace=True, level=1)
total_df.reset_index(inplace=True)
total_df.dropna(inplace=True, how="any")

total_df.set_index(["idm", "assignment"], inplace=True)
```

```python
total_df = total_df.unstack("assignment")
```

```python
result = pd.DataFrame(total_df.count(axis=1), columns=["passed"])
```

```python
result
```

```python
# add new column where called bonus where 0.3 is put in where 5 assignments are passed and 0.7 if 6 assignments are passed
result["bonus"] = result["passed"].apply(lambda x: 0.33 if x == 5 else 0.66 if x == 6 else 0)
```

```python
result.to_excel(exercise_path / "result_final.xlsx")
```
