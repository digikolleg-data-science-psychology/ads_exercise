```python
from pathlib import Path
```

```python
assignment = "2-python-basics-II"
file_name = "2_python_basics_II.ipynb"
```

```python
base_path = Path("course_directory/cloned_repos")
```

```python
# search for all wrongly named files (include subfolders)
for file in base_path.joinpath(assignment).rglob("*.ipynb"):
    if file.is_file() and file.name != file_name:
        
        # print file
        print(file)

        # rename file
        new_file = file.parent.joinpath(file_name)
        file.rename(new_file)

        # do git commit and push in parent directory (test this first!)
        !cd {file.parent} && git add {file_name} && git commit -m "rename {file.name} to {file_name}" && git push
```
